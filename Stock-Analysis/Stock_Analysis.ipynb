```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```
data = pd.read_csv('satma.txt', delimiter= '\s+', index_col=False)
data
```

```
data = data.to_numpy()
data
```

```
Base_volume = data[:,1]
shopping_line = data[:,2]
Sales_volume = data[:,3]
time = data[:,4]
```

```
%matplotlib inline
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=[10,10])  # create a plot figure
#time = np.linspace(0, 10, 100)
# create the first of two panels and set current axis
#ax = plt.axes()
plt.plot(time, Sales_volume, label="Sales_volume", color='k');
plt.plot(time, Base_volume, '--', label="Base_volume", color='r',)
#plt.bar(time, shopping_line, label="Blue Bar", color='b')
plt.plot(time, shopping_line,':b', label="shopping_line", color='b');
#ax.set(xlim=(0, 10), ylim=(-2, 2),
#       xlabel='time', ylabel='sin(time)',
#       title='Stock-Analysis');

plt.xlabel("Time")
plt.ylabel("Volume")
plt.title("Stock-Analysis")
plt.legend()
#plt.grid(True,color='k')
plt.show()
```
